````python
%%tutor -l java -k

class Punto {
   
    private double x;
    private double y;

    public Punto() {
        x = 0;
        y = 0;
    }

    public double getX() {
        return x;
    }

    public void setX(double x) {
        this.x = x;
    }

    public double getY() {
        return y;
    }

    public void setY(double y) {
        this.y = y;
    }
   
    public void setXY(double x, double y) {
        this.x = x;
        this.y = y;
    }

    public Punto(double x, double y) {
        this.x = x;
        this.y = y;
    }
   
    public void imprimirCoordenadas() {
        System.out.println("(" + x + ", " + y +")");

    }
   
    public double calcularDistancia(double x, double y) {
        double d;
        d = Math.sqrt(Math.pow(this.x - x, 2) +
                      Math.pow(this.y - y, 2));
        return d;  
    }
   
    public double calcularDistancia(Punto p) {
        double d;
        d = Math.sqrt(Math.pow(this.x - p.x, 2) +
                      Math.pow(this.y - p.y, 2));
        return d;
    }
   
}

public class PointDemo {
    public static void main(String[] args) {
        /* Creando 4 objetos tipo punto */
        Punto O = new Punto();
        Punto A = new Punto(2,3);
        Punto B = new Punto(-3,1);
        Punto C = new Punto(-1.5,-2.5);
        
        /* Imprimiendo las coordenadas */
        System.out.print("O: ");
        O.imprimirCoordenadas();
        System.out.print("A: ");
        A.imprimirCoordenadas();
        System.out.print("B: ");
        B.imprimirCoordenadas();
        System.out.print("C: ");
        C.imprimirCoordenadas();
        
        /* Imprimiendo distancias */
        // Distancia entre A y B
        double d_AB = A.calcularDistancia(B.getX(),B.getY());
        System.out.println("Distancia (A,B): " + d_AB);
        // Distancia entre B y C
        double d_BC;
        d_BC = B.calcularDistancia(C);
        System.out.println("Distancia (B,C): " + d_BC);
    }
}

````

## Array de objetos ##
Es posble crear arreglos de objetos que son instancias de las clases que ha sido escrita. La siguiente forma muestra la sintaxis empleada para declarar un arra de objetos:

```java
nombreClase []nombreArray = new nombreClase[sizeArray];
```

Cuando un array de Objetos es instanciado, cada elemento del arreglo es inicializado a **null**. **null** es un valor especial que indica que los elementos del array aun no han sido referenciados con objetos. Para hacer que cada referenciar cada elemento es necesario clear objetos indiviualmente tal y como se muestra a continuación:

```java
nombreArray[0] = new nombreClase(parametros,...);
nombreArray[1] = new nombreClase(parametros,...);
...
nombreArray[i] = new nombreClase(parametros,...);
...
nombreArray[N] = new nombreClase(parametros,...);
```

**Ejemplo**: Crear un arreglo de 4 puntos con las siguientes coordenadas: A(2,1), B(-3,4), C(5,-6), D(-7,-8).

````python
%%tutor -l java -k

class Punto {
   
    private double x;
    private double y;

    public Punto() {
        x = 0;
        y = 0;
    }

    public double getX() {
        return x;
    }

    public void setX(double x) {
        this.x = x;
    }

    public double getY() {
        return y;
    }

    public void setY(double y) {
        this.y = y;
    }
   
    public void setXY(double x, double y) {
        this.x = x;
        this.y = y;
    }

    public Punto(double x, double y) {
        this.x = x;
        this.y = y;
    }
   
    public void imprimirCoordenadas() {
        System.out.println("(" + x + ", " + y +")");

    }
   
    public double calcularDistancia(double x, double y) {
        double d;
        d = Math.sqrt(Math.pow(this.x - x, 2) +
                      Math.pow(this.y - y, 2));
        return d;  
    }
   
    public double calcularDistancia(Punto p) {
        double d;
        d = Math.sqrt(Math.pow(this.x - p.x, 2) +
                      Math.pow(this.y - p.y, 2));
        return d;
    }
   
}

public class PointDemo2 {
    public static void main(String[] args) {
        /* Creando 4 objetos tipo punto */
        final int NUMPUNTOS = 4;
        Punto[] puntos;
        puntos = new Punto[NUMPUNTOS];
        Punto A = new Punto(2,1);
        Punto B = new Punto(-3,4);
        puntos[0] = A;
        puntos[1] = B;
        puntos[2] = new Punto(5,-6);
        puntos[3] = new Punto(-7,-8); 
        imprimirArray(puntos);
        
    }
    
    public static void imprimirArray(Punto []v) {
        for(Punto e: v) {
            e.imprimirCoordenadas();
        }
    }
}
````
